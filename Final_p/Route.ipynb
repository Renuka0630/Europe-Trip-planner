```python
import pandas as pd
import folium
from folium import plugins
from itertools import permutations
from scipy.spatial import distance_matrix
```

```python
# Load the data
data = pd.read_excel('C:\\Users\\Ashwin Rathod\\Documents\\r_project\\Map Coordinates.xlsx')
data
```

```python
# Create a map centered at France
map_center = [46.6034,	1.8883]  # France's coordinates
m = folium.Map(location=map_center, zoom_start=5)
```

```python
# Add markers for each country
for index, row in data.iterrows():
    folium.Marker([row['LATITUDE'], row['LONGITUDE']], tooltip=row['COUNTRY']).add_to(m)
```

```python
m
```

```python
# Specify the mandatory countries
mandatory_countries = ['FRANCE', 'ITALY', 'SWITZERLAND', 'UNITED KINGDOM']
```

```python
# Find the indices of the mandatory countries
mandatory_indices = [data.index[data['COUNTRY'] == country].tolist()[0] for country in mandatory_countries]
```

```python
# Calculate distance matrix
dist_matrix = distance_matrix(data[['LATITUDE', 'LONGITUDE']], data[['LATITUDE', 'LONGITUDE']])
```

```python
# Find the combination of 3 additional countries that minimizes the total distance
min_distance = float('inf')
best_route = None
for perm in permutations([idx for idx in range(len(data)) if idx not in mandatory_indices], 3):
    perm_distance = sum([dist_matrix[mandatory_indices[0], perm[0]]] + [dist_matrix[perm[i], perm[i + 1]] for i in range(2)] + [dist_matrix[perm[-1], mandatory_indices[1]]])
    if perm_distance < min_distance:
        min_distance = perm_distance
        best_route = perm
```

```python
# Add the mandatory countries to the route
route_indices = [mandatory_indices[0]] + list(best_route) + mandatory_indices[1:]

```

```python
# Add a polyline for the route
route_coordinates = [[data.iloc[i]['LATITUDE'], data.iloc[i]['LONGITUDE']] for i in route_indices]
folium.PolyLine(route_coordinates, color='blue', weight=2.5, opacity=1).add_to(m)
```

```python
# Get the names of the selected countries
selected_countries = [data.iloc[i]['COUNTRY'] for i in route_indices]
```

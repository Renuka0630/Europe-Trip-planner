#Maximum Hostel Price in different Countries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file
file_path = r"C:\Users\Ashwin Rathod\Documents\r_project\Euro Project Data.xlsx"
df = pd.read_excel(file_path)
```

```python
# Convert 'HOSTEL PRICE' to numeric
df['HOSTEL PRICE'] = pd.to_numeric(df['HOSTEL PRICE'], errors='coerce')

# Plot a bar chart of hostel prices by country
plt.figure(figsize=(12, 8))
sns.barplot(x='COUNTRY', y='HOSTEL PRICE', data=df, estimator=max)
plt.title('Maximum Hostel Price by Country')
plt.xticks(rotation=45)
plt.show()
```

#Per Day Spend in European Cities

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
df = pd.read_excel('C:\\Users\\Ashwin Rathod\\Documents\\r_project\\Euro Project Data.xlsx')

# Group the data by CITY and sum the PRICE
city_costs = df.groupby('CITY')['PRICE'].sum().reset_index()

# Sort the data by total cost
city_costs_sorted = city_costs.sort_values(by='PRICE', ascending=False)

# Plot
plt.figure(figsize=(14, 8))
plt.bar(city_costs_sorted['CITY'], city_costs_sorted['PRICE'], color='skyblue')
plt.title('Total Cost Comparison by City')
plt.xlabel('City')
plt.ylabel('Total Cost')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#Food and SughtSeeing

```python
import pandas as pd
import os

# Load the data from the Excel file
df = pd.read_excel(r'C:\Users\Ashwin Rathod\Documents\r_project\Euro Project Data.xlsx')

def get_city_info(city_name):
    """Function to retrieve sightseeing locations and food in a given city."""
    city_data = df[df['CITY'] == city_name]
    if city_data.empty:
        return f"No information found for {city_name}"
    else:
        sightseeing_locations = city_data.iloc[0]['SIGHTSEEING']
        food_options = city_data.iloc[0]['FOOD']
        return f"\033[1;37mSightseeing locations in {city_name}:\n{sightseeing_locations}\nFood options in {city_name}:\n{food_options}\033[0m"




```

```python

# Example usage
city_name = 'Paris'  # Change this to the desired city name
city_info = get_city_info(city_name)
print(city_info)
```

#Transportation From Cities and Transportation Medium

```python
import pandas as pd
import os

# Load the data from the Excel file
df = pd.read_excel(r'C:\Users\Ashwin Rathod\Documents\r_project\Euro Project Data.xlsx')

def get_city_info(city_name):
    """Function to retrieve sightseeing locations, food, transport, and price/time in a given city."""
    city_data = df[df['CITY'] == city_name]
    if city_data.empty:
        return f"No information found for {city_name}"
    else:
        transport_info = city_data.iloc[0]['TRANSPORT']
        price_time = city_data.iloc[0]['PRICE / TIME']
        return f"\033[1;37mTransport in {city_name}:\n{transport_info}\nPrice/Time in {city_name}:\n{price_time}\033[0m"



```

```python
# Example usage
city_name = 'Paris'  # Change this to the desired city name
city_info = get_city_info(city_name)
print(city_info)
```
